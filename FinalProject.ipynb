#### Imports

```python
from xml.etree.ElementTree import ElementTree as et
from xml.etree.ElementTree import fromstring
import urllib.request
import bs4 as bs
```

#### Reuters XML Feed

```python
source = urllib.request.urlopen('http://feeds.reuters.com/reuters/topNews').read()
```

```python
tree = fromstring(source.decode('utf-8'))
```

```python
root = tree.getchildren()
firstchild = root[0]
```

```python
itemlist = firstchild.findall('item')
len(itemlist)
for items in itemlist:
    print("Title: ",items[0].text,"\n")
    #source = urllib.request.urlopen(items[2].text).read()
    #soup = bs.BeautifulSoup(source,'lxml')
    #story = soup.find_all("p", {"class": "StandardArticleBody_body"})
    print(itemlist.index(items)+1,items[2].text,"\n")
    #print("Body text : ",source,"\n")
```
